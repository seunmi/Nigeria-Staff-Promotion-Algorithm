### Data cleaning

```python
df.drop(['EmployeeNo','State_Of_Origin'],axis=1,inplace=True)
dftest.drop(['EmployeeNo','State_Of_Origin'],axis=1,inplace=True)

```

```python
df.drop('Year_of_recruitment',axis=1,inplace=True)
dftest.drop('Year_of_recruitment',axis=1,inplace=True)

```

```python
df['Qualification']=df['Qualification'].fillna('First Degree or HND')
dftest['Qualification']=dftest['Qualification'].fillna('First Degree or HND')
```

```python
dummy_variable=df.select_dtypes(['object'])
dummy_variable2=dftest.select_dtypes(['object'])

```

```python
dummy_variable.head()
```

```python
train_dummy=pd.get_dummies(dummy_variable,drop_first=True)
test_dummy=pd.get_dummies(dummy_variable2,drop_first=True)
```

```python
train_dummy.head()
```

```python
test_dummy.head()
```

```python
df=df.drop(list(dummy_variable.columns),axis=1)
dftest=dftest.drop(list(dummy_variable2.columns),axis=1)
```

```python
df=pd.concat((df,train_dummy),axis=1)
dftest=pd.concat((dftest,test_dummy),axis=1)
```

### Training the model using XGboost

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
model = XGBClassifier(random_state=0, n_jobs=-1,learning_rate=0.1,n_estimators=100,max_depth=8)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
predictions = [round(value) for value in y_pred]
```

```python
from sklearn.metrics import  confusion_matrix
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True,fmt='d')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
model.score(x_test,y_test)
```
